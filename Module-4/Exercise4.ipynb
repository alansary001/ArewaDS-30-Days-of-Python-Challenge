## Module 4 exercises (Strings)

```python
## 1. Concatenation of strings
str1 = "Thirty"
str2 = "Days"
str3 = "Of"
str4 = "Python"
space = " "
ConcatenatedString = str1 + space + str2 + space + str3 + space + str4
print(ConcatenatedString)

str5 = "Coding"
str6 = "For"
str7 = "All"
ConcatenatedString2 = str5 + space + str6 + space + str7    
print(ConcatenatedString2)
company = ConcatenatedString2
print (company)

length = len(company)
print (length)
print (company.upper())
print (company.lower())
print (company.capitalize())
print (company.title())
print (company.swapcase())

first_word = company.split()[0]
print(first_word)

# Check if the string contains the word 'Coding'
print(company.find('Coding'))  # Output: 0 (position of 'Coding')

# Replace 'coding' with 'Python'
company_replaced = company.replace('Coding', 'Python')
print(company_replaced)  # Output: Python For All

# Change 'Python For Everyone' to 'Python For All'
sentence = 'Python For Everyone'
new_sentence = sentence.replace('Everyone', 'All')
print(new_sentence)  # Output: Python For All

# Split the string 'Coding For All' using space as separator
split_str = company.split()
print(split_str)  # Output: ['Coding', 'For', 'All']

# Split the string 'Facebook, Google, Microsoft, Apple, IBM, Oracle, Amazon' at the comma
companies = "Facebook, Google, Microsoft, Apple, IBM, Oracle, Amazon"
split_companies = companies.split(', ')
print(split_companies)

# Character at index 0
print(company[0])  # Output: C

# Last index of the string
print(len(company) - 1)  # Output: 14

# Character at index 10
print(company[10])  # Output: A

# Create acronym for 'Python For Everyone'
acronym_pfe = ''.join([word[0] for word in 'Python For Everyone'.split()])
print(acronym_pfe)  # Output: PFE

# Create acronym for 'Coding For All'
acronym_cfa = ''.join([word[0] for word in 'Coding For All'.split()])
print(acronym_cfa)  # Output: CFA

# Position of the first occurrence of 'C' in 'Coding For All'
print(company.index('C'))  # Output: 0

# Position of the first occurrence of 'F' in 'Coding For All'
print(company.index('F'))  # Output: 7

# Use rfind to find the last occurrence of 'l' in 'Coding For All People'
print(company.rfind('l'))  # Output: 13

# Find the first occurrence of 'because' in the sentence
sentence2 = 'You cannot end a sentence with because because because is a conjunction'
print(sentence2.find('because'))  # Output: 38

# Use rindex to find the position of the last occurrence of 'because'
print(sentence2.rindex('because'))  # Output: 46

# Slice out 'because because because'
print(sentence2[38:69])  # Output: because because because

# Does 'Coding For All' start with 'Coding'?
print(company.startswith('Coding'))  # Output: True

# Does 'Coding For All' end with 'coding'?
print(company.endswith('coding'))  # Output: False

# Remove left and right trailing spaces
trailing_spaces = '   Coding For All      '
print(trailing_spaces.strip())  # Output: 'Coding For All'

# Check if a variable is a valid identifier
print("30DaysOfPython".isidentifier())  # Output: True
print("thirty_days_of_python".isidentifier())  # Output: True

# Join the list of Python libraries with a hash and space
libraries = ['Django', 'Flask', 'Bottle', 'Pyramid', 'Falcon']
print(" # ".join(libraries))  # Output: Django # Flask # Bottle # Pyramid # Falcon

# Use new line escape sequence
print("I am enjoying this challenge.\nI just wonder what is next.")

# Use tab escape sequence to format data
print("Name\tAge\tCountry\tCity")
print("Asabeneh\t250\tFinland\tHelsinki")

# Use string formatting for circle area
radius = 10
area = 3.14 * radius ** 2
print(f"The area of a circle with radius {radius} is {area} meters square.")

# Use string formatting for mathematical operations
print(f"8 + 6 = {8 + 6}")
print(f"8 - 6 = {8 - 6}")
print(f"8 * 6 = {8 * 6}")
print(f"8 / 6 = {8 / 6:.2f}")
print(f"8 % 6 = {8 % 6}")
print(f"8 // 6 = {8 // 6}")
print(f"8 ** 6 = {8 ** 6}")



```
