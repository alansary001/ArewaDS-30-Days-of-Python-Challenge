```python
## Function to Add Two Numbers
def add(x, y):
    return x + y

## function for area of circle
def area_of_circle(radius):
    return 3.14159 * radius ** 2

## Sum of Arbitrary Numbers
def sum_of_numbers(*args):
    return sum(args)

## Convert °C to °F
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

##  Check Season
def check_season(month):
    month = month.lower()
    if month in ['september', 'october', 'november']:
        return "Autumn"
    elif month in ['december', 'january', 'february']:
        return "Winter"
    elif month in ['march', 'april', 'may']:
        return "Spring"
    elif month in ['june', 'july', 'august']:
        return "Summer"
    return "Invalid month"

## Calculate Slope
def calculate_slope(x1, y1, x2, y2):
    return (y2-y1)/(x2-x1)

## Solve Quadratic Equation
def solve_quadratic_eqn(a, b, c):
    d = b**2 - 4*a*c
    if d > 0:
        x1 = (-b + d**0.5) / 2*a
        x2 = (-b - d**0.5) / 2*a
        return x1, x2
    elif d == 0:
        x = -b / 2*a
        return x
    else:
        return "No Real Roots"
    

## print list
def print_list(lst):
    for i in lst:
        print(i)

## Reverse a String
def reverse_string(string):
    return string[::-1]

## Capitalize List Items
def capitalize_list_items(lst):
    return [i.capitalize() for i in lst]

## Add Item to List
def add_item(lst, item):
    lst.append(item)
    return lst

## Remove Item from List
def remove_item(lst, item):
    lst.remove(item)
    return lst

## Sum of Numbers in a Range
def sum_of_numbers_in_a_range(start, end):
    return sum(range(start, end+1))

## Sum of Odds in a Range
def sum_of_odds_in_a_range(start, end):
    return sum(i for i in range(start, end+1) if i % 2 != 0)

## Sum of Evens in a Range
def sum_of_evens_in_a_range(start, end):
    return sum(i for i in range(start, end+1) if i % 2 == 0)









```

```python
## Evens and Odds Counter
def evens_and_odds_counter(lst):
    evens = len([i for i in lst if i % 2 == 0])
    odds = len([i for i in lst if i % 2 != 0])
    return evens, odds

## Factorial
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

## Check if Variable is Empty
def is_empty(variable):
    if variable == "":
        return True
    return False

##Statistical Functions
import statistics

def calculate_mean(lst):
    return statistics.mean(lst)

def calculate_median(lst):
    return statistics.median(lst)

def calculate_mode(lst):
    return statistics.mode(lst)

def calculate_range(lst):
    return max(lst) - min(lst)

def calculate_variance(lst):
    return statistics.variance(lst)

def calculate_std(lst):
    return statistics.stdev(lst)


```

```python
## Check Prime Number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

## Check Unique Items in List
def has_duplicates(lst):
    return len(lst) != len(set(lst))

## Check Uniform Data Types in List
def is_homogenous(lst):
    return len(set(map(type, lst)))

## Check Valid Python Variable
def is_valid_variable(variable):
    return variable.isidentifier()

## Most Spoken Languages
def most_spoken_languages(data, top_n=10):
    language_count = {}
    for country in data:
        for lang in country['languages']:
            language_count[lang] = language_count.get(lang, 0) + 1
    sorted_languages = sorted(language_count.items(), key=lambda x: x[1], reverse=True)
    return sorted_languages[:top_n]

## Most Populated Countries
def most_populated_countries(data, top_n=10):
    sorted_countries = sorted(data, key=lambda x: x['population'], reverse=True)
    return sorted_countries[:top_n]



```
